```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customer_transactions_final.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.describe()
```

```python
df["purchase_amount"].sum()
```

```python
df["category"].value_counts()
```

```python
df.groupby("gender")["purchase_amount"].sum()
```

```python
df.groupby("season")["purchase_amount"].sum()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="category", data=df)
plt.xticks(rotation=45)
plt.title("Number of Purchases by Category")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="gender", y="purchase_amount", data=df)
plt.title("Purchase Amount by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df["payment_method"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df["age"], bins=20)
plt.title("Customer Age Distribution")
plt.show()
```
